```python
import matplotlib.pyplot as plt

diseases = ["Atelectasis", "Cardiomegaly", "Effusion", "Infiltration", "Mass", "Nodule",
            "Pneumonia", "Pneumothorax", "Consolidation", "Edema", "Emphysema",
            "Fibrosis", "Pleural Thickening", "Hernia", "Overall (mAUC)"]
index = range(len(diseases))

def create_tta_barplot(index, auc_with_tta, auc_without_tta, diseases, model_name):
    fig, ax = plt.subplots(figsize=(12, 8))

    # Colors specified by the user
    colors = ['#FFAFCC', '#A2D2FF']

    # Create horizontal bar plots
    bar_width = 0.35
    rects1 = ax.barh(index, auc_with_tta, bar_width, color=colors[0], edgecolor='grey', label='AUC w/ TTA')
    rects2 = ax.barh([p + bar_width for p in index], auc_without_tta, bar_width, color=colors[1], edgecolor='grey', label='AUC w/o TTA')

    # Add labels, title, and custom y-axis tick labels
    ax.set_ylabel('Disease')
    ax.set_xlabel('AUC scores')
    ax.set_title(f'AUC scores given class with and without TTA for {model_name}', fontsize=24)
    ax.set_yticks([p + bar_width / 2 for p in index])
    ax.set_yticklabels(diseases)
    ax.legend()

    # Show the plot
    plt.tight_layout()
    plt.savefig(f'{model_name.lower()}_tta_barplot.png')


```
