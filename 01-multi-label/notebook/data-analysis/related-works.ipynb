```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, Normalize

# Data setup
data = {
    'Technique': ['Scale', 'Center Crop', 'ImageNet', 'Rotation', 'Brightness', 'Resize', 'Horizontal Flip', 'Resized Crop', 'Color Jitter'],
    'Count': [1, 2, 1, 2, 3, 4, 5, 4, 2]
}
df = pd.DataFrame(data)

# Custom colors setup
custom_colors = ["#FFAFCC","#FFAFCC", "#FFC8DD","#FFC8DD", "#CDB4DB","#CDB4DB", "#A2D2FF", "#BDE0FE","#EBF5FF"]

# Plotting the bar chart
plt.figure(figsize=(7, 4))
bars = plt.bar(df['Technique'], df['Count'], color=custom_colors,  edgecolor='grey')
plt.title('Frequency of Augmentation Techniques')
plt.xlabel('Augmentation Technique')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("freq-augmentations.png")
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Data setup
optimizers = ['Adam', 'AdamW', 'SGD']
counts = [3, 4, 1]  # Count of each optimizer based on the bar heights

# Color setup based on the chart, using similar shades of purple and pink
colors = ['#FFAFCC', '#CDB4DB', '#BDE0FE']

# Plotting the bar chart
plt.figure(figsize=(7, 4))
plt.bar(optimizers, counts, color=colors, edgecolor='grey')
plt.title('Frequency of Optimizers')
ax = plt.gca()
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.xlabel('Optimizer')
plt.ylabel('Count')
plt.ylim(0, 4.5)  # Set the limit to match the maximum count visible on the chart
plt.tight_layout()
plt.savefig("optimizers.png")
```
